```python
import pandas as pd
import numpy as np

tt_dict = pd.read_excel (r'C:\Users\trainor-gilhamt\Documents\Unilever Thailand\Fabric\Fabric Enhancer_by Channel & Tops_Nielsen.xlsx (edited).xlsx', sheet_name = None, header = [0])

tt = pd.DataFrame()
for k in tt_dict.keys():
    tempdf = tt_dict[k].copy()
    tempdf['Metrics'] = k
    tt = tt.append(tempdf, sort = False)
    
tt
```

```python
col_name="Metrics"
first_col = tt.pop(col_name)

tt.insert(0, col_name, first_col)

tt
```

```python
#tt2 = tt.drop(columns=['Category','Manufacturer', 'Brand', 'Segment', 'Subsegment','Variant','Subvariant','Basesize','Packtype'])
#tt2
```

```python
Columns = ['LDESC', 'EAN']
tt.loc[:,Columns] = tt.loc[:,Columns].ffill()
tt
```

```python
Raw_Long = tt.set_index([('LDESC'),
            ('EAN'),
            ('Metrics'),

            ]).stack(dropna=False).reset_index().rename(columns = {'level_3': 'Date' , 0: 'Data'})
         
Raw_Long
```

```python
Raw_Long['Date'] = pd.to_datetime(Raw_Long['Date'])
```

```python
dataTypeSeries = Raw_Long.dtypes
print(dataTypeSeries)
```

```python
for col in Raw_Long:
    print(Raw_Long[col].unique())
```

```python
Raw_Long['Metrics'].replace({'Value Sales':'Sales', 'Wtd Selling Dist':'ACV', 'Unit Sales':'Units' }, inplace=True)
Raw_Long
```

```python
# Raw_Long.set_index(['Retailer', 'UPC', 'Prod Desc', 'Date'], append=True)
```

```python
import numpy as np
 
Raw_Long['Index'] = np.arange(len(Raw_Long))
print (Raw_Long)
```

```python
Raw_Long_2 = Raw_Long.pivot(index= ['Index', 'LDESC' , 'EAN' , 'Date'] , columns= 'Metrics' , values='Data')

Raw_Long_2
```

```python
Raw_Long_2.reset_index(inplace=True)
Raw_Long_2
```

```python
Raw_Long_2['EAN'] = Raw_Long_2['EAN'].astype('str').replace('\.0', '', regex=True)
Raw_Long_2
```

```python
Final = Raw_Long_2.drop(columns=['LDESC','Index'])
Final
```

```python
Final.fillna(0, inplace = True)
Final
```

```python
Final2 = Final.rename(columns={'EAN':'UPC_DESC', 'Date':'WEEK', 
                                   'Units':'UNITS', 'Sales':'SALES'})
Final2
```

```python
# groupby UPC, Retailer, UPC_DESC, 
# Choose max ACV

Final2a = Final2.groupby(['UPC_DESC', 'WEEK']).agg({ 'UNITS':['sum'] , 'SALES': ['sum'] , 'ACV':['max']})
Final2a
```

```python
Final2a.reset_index(inplace=True)
Final2a
```

```python
Final2a.columns = Final2a.columns.droplevel(-1)
Final2a
```

```python
Final3 = Final2a.loc[(Final2a['WEEK'] >= '2020-05-01') & (Final2a['WEEK'] <= '2022-05-01' )]
Final3
```

```python
dataTypeSeries = Final3.dtypes
print(dataTypeSeries)
```

```python
Final4 = Final3.loc[(Final3['ACV'] > 0) & (Final3['UNITS'] > 0) & (Final3['SALES'] > 0)].reset_index(drop=True)
Final4
```

```python
Final4['UPC_DESC'] = Final4['UPC_DESC'].astype('str').replace('\.0', '', regex=True)
Final4
```

```python
Final4["RETAILER"] = "TOPS"
Final4
```

```python
Final4 = Final4[['RETAILER', 'UPC_DESC', 'WEEK', 'UNITS', 'SALES', 'ACV']]
Final4
```
