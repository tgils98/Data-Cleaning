```python
import glob
import pandas as pd



df_list = []
for file in glob.glob(r'C:\Users\trainor-gilhamt\Documents\Unilever NL\Retailer Raw Data\*.xlsx'):
# Tweak this to work for your actual filepaths, if needed.
    FN, Retailer = file.split('.')[0].split('-')[1:]
    df = (pd.read_excel(file)

          .assign(Folder=FN, Retailer=Retailer))

    df_list.append(df)


    df = pd.concat(df_list)
```

```python
df
```

```python
col_name="Retailer"
first_col = df.pop(col_name)

df.insert(0, col_name, first_col)

df
```

```python
Columns = ['Prod Desc', 'UPC']
df.loc[:,Columns] = df.loc[:,Columns].ffill()
df
```

```python
df2 = df.drop(columns=['Folder'])
```

```python
Raw_Long = df2.set_index([('Retailer'),
            ('UPC'),
            ('Prod Desc'),
            ('Facts'),

            ]).stack(dropna=False).reset_index().rename(columns = {'level_4': 'Date' , 0: 'Data'})
         
Raw_Long
```

```python

Raw_Long['Date'] = pd.to_datetime(Raw_Long['Date'])
```

```python
dataTypeSeries = Raw_Long.dtypes
print(dataTypeSeries)
```

```python
# Raw_Long.set_index(['Retailer', 'UPC', 'Prod Desc', 'Date'], append=True)
```

```python
import numpy as np
 
Raw_Long['Index'] = np.arange(len(Raw_Long))
print (Raw_Long)
```

```python
Raw_Long_2 = Raw_Long.pivot(index= ['Index', 'Retailer' , 'UPC' , 'Prod Desc' , 'Date'] , columns= 'Facts' , values='Data')

Raw_Long_2
```

```python
Raw_Long_2.reset_index(inplace=True)
Raw_Long_2
```

```python
Final = Raw_Long_2.drop(columns=['Index'])
Final
```

```python
Final.fillna(0, inplace = True)
Final
```

```python
Final2 = Final.rename(columns={'Retailer':'RETAILER','UPC':'UPC_DESC','Prod Desc':'UPC_SHORT_DESC', 'Date':'WEEK', 
                                   'Units':'UNITS', 'Value (in EUR)':'SALES (EUR)', 'Weight. Distr.':'ACV'})
Final2
```

```python
# groupby UPC, Retailer, UPC_DESC, 
# Choose max ACV

Final2a = Final2.groupby(['RETAILER' , 'UPC_DESC' , 'UPC_SHORT_DESC' , 'WEEK']).agg({ 'UNITS':['sum'] , 'SALES (EUR)': ['sum'] , 'ACV':['sum']})
Final2a
```

```python
Final2a.reset_index(inplace=True)
Final2a
```

```python
Final3 = Final2.rename(columns={'UNITSsum':'UNITS', 'SALES (GBP)sum':'SALES (GBP)', 'ACVsum':'ACV'})
Final3
```

```python
Final3b = Final3.loc[(Final3['WEEK'] >= '2020-08-23') & (Final3['WEEK'] <= '2022-08-21' )]
Final3b
```

```python
dataTypeSeries = Final3b.dtypes
print(dataTypeSeries)
```

```python
a = Final3b[(Final3b['UNITS']!=0) | (Final3b['SALES (EUR)']!=0) | (Final3b['ACV']!=0)]
```
