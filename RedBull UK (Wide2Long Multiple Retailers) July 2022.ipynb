```python
import pandas as pd
import numpy as np

sales_dict = pd.read_excel (r'C:\Users\trainor-gilhamt\Documents\Red Bull UK\RedBull Value Sales MAT July 2022.xlsx', sheet_name = None, header = [0])

sales = pd.DataFrame()
for k in sales_dict.keys():
    tempdf = sales_dict[k].copy()
    tempdf['Retailer'] = k
    sales = sales.append(tempdf, sort = False)
sales
```

```python
import pandas as pd
import numpy as np

units_dict = pd.read_excel (r'C:\Users\trainor-gilhamt\Documents\Red Bull UK\RedBull Unit Sales MAT July 2022.xlsx', sheet_name = None, header = [0])

units = pd.DataFrame()
for k in units_dict.keys():
    tempdf = units_dict[k].copy()
    tempdf['Retailer'] = k
    units = units.append(tempdf, sort = False)

units
```

```python
import pandas as pd
import numpy as np

dist_dict = pd.read_excel (r'C:\Users\trainor-gilhamt\Documents\Red Bull UK\RedBull Dist MAT July 2022.xlsx', sheet_name = None, header = [0])

dist = pd.DataFrame()
for k in dist_dict.keys():
    tempdf = dist_dict[k].copy()
    tempdf['Retailer'] = k
    dist = dist.append(tempdf, sort = False)

dist
```

```python
Merged = units.merge(sales,how ='left').merge(dist,how ='left')

Merged
```

```python
Merged['Unit Sales MAT'] = Merged['Unit Sales MAT'].fillna(0)
```

```python
Merged['Value Sales MAT'] = Merged['Value Sales MAT'].fillna(0)
```

```python
Merged['Dist MAT'] = Merged['Dist MAT'].fillna(0)
```

```python
Merged
```

```python
Final = Merged.rename(columns={'INFOITEM':'UPC', 'Unit Sales MAT':'Units', 'Dist MAT':'ACV', 'Value Sales MAT':'Sales'})


column_names = ["Retailer", "UPC", "Date", "Units", "Sales", "ACV"]

Final = Final.reindex(columns=column_names)

Final
```

```python
Final.to_csv(r'C:\Users\trainor-gilhamt\Documents\Red Bull UK\RM Formatted Data\RedBull_Curve_Model_Data_Transformed.CSV', index = False)
```
