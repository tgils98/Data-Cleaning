```python
df2 = pd.read_csv(r'C:\Users\trainor-gilhamt\Documents\Ferrero UK\Ferrero UK Spreads Curve\Ferrero_UK_Spreads_Curve_Model_Data_Transformed 2.csv')

df2
```
